```python
year=range(2000,2008)
apples=[0.895,0.91,0.919,0.926,0.929,0.931]
oranges=[0.962,0.941,0.930,0.918,0.908,0.907]

```

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel("year")
plt.ylabel("yield")
plt.title("apples and oranges yield / per hectare")
plt.legend(["apples","oranges"])
```

```python
plt.plot(years,apples,marker='o')
plt.plot(years,oranges,marker='x')
plt.xlabel("year")
plt.ylabel("yield")
plt.title("apples and oranges yield / per hectare")
plt.legend(["apples","oranges"])
```

```python
plt.plot(years,apples,marker='s',c='b',ls='-',ms=8,mew=2,mec='navy')
plt.plot(years,oranges,marker='o',c='r',ls='--',lw=3,ms=10,alpha=0.5)
plt.xlabel("year")
plt.ylabel("yield")
plt.title("apples and oranges yield / per hectare")
plt.legend(["apples","oranges"])
```

```python
#if no line is specified in m=fmt, only markers aew drawn
plt.plot(years,apples,'or')
plt.plot(years,oranges,'or')
```

```python
plt.figure(figsize=(4,4))
plt.plot(years,apples,'or')
plt.title('yield of oranges')
```
